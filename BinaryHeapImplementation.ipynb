```python
class minHeap:
    def __init__(self,maxcapacity):
        self.currcapacity = 0
        self.maxcapacity = maxcapacity
        self.value = [-1]*maxcapacity
        
    def isEmpty(self) -> bool:
        return self.currcapacity == 0
        
    def getParentIndex(self, childIndex):
        return (childIndex - 1)//2

    def getLeftChildIndex(self, parentIndex):
        return (2*parentIndex +1)

    def getRightChildIndex(self, parentIndex):
        return (2*parentIndex +2)
    
    def hasParent(self, childIndex):
        return (childIndex != 0) and (self.getParentIndex(childIndex) >= 0)

    def hasLeftChild(self, parentIndex):
        return self.getLeftChildIndex(parentIndex) < self.currcapacity

    def hasRightChild(self, parentIndex):
        return self.getRightChildIndex(parentIndex) < self.currcapacity

    def isLeafNode(self, index):
        if self.hasLeftChild(index):
            return False
        return True

    def swap(self, parent, child):
        self.value[parent], self.value[child] = self.value[child], self.value[parent]

    def insertWithPriority(self, val):
        if self.isEmpty():
            self.value[0] = val
            self.currcapacity = self.currcapacity + 1
            return
        self.value[self.currcapacity] = val
        self.currcapacity = self.currcapacity + 1
        self.bubbleUp()

    def pullHighestPriority(self):
        if self.isEmpty():
            return
        returnValue = self.value[0]
        lastIndex = self.currcapacity - 1
        self.value[0] = self.value[lastIndex]
        self.value[lastIndex] = -1
        self.currcapacity = self.currcapacity - 1
        self.bubbleDown()
        return returnValue

    def peek(self):
        if self.isEmpty():
            return
        return self.value[0]
    
    def bubbleDown(self):
        parentIndex = 0
        while (self.isLeafNode(parentIndex) == False):
            childMin = self.getLeftChildIndex(parentIndex)
            if self.hasRightChild(parentIndex) and (self.value[self.getRightChildIndex(parentIndex)] < self.value[self.getLeftChildIndex(parentIndex)]):
                childMin = self.getRightChildIndex(parentIndex)
            if self.value[parentIndex] < self.value[childMin]:
                 break #Min already at parent
            else:
                 self.swap(parentIndex, childMin)
            parentIndex = childMin
    
    def bubbleUp(self):
        currvalpos = self.currcapacity - 1
        while self.hasParent(currvalpos) and (self.value[currvalpos] < self.value[self.getParentIndex(currvalpos)]):
            self.swap(self.getParentIndex(currvalpos), currvalpos)
            currvalpos = self.getParentIndex(currvalpos)
```
